### 3.	Create an 1D array of 100 evenly spaced numbers between 0 and 100 using NumPy and save it to variable X.
    a.	Create a plot using plt.subplots() and plot X versus X2
    b.	Create a scatter plot of X versus the exponential of X (np.exp(X))


```python
X = np.linspace(0, 100, 100)
X_squared = X ** 2
```

```python
fig, ax1 = plt.subplots()
ax1.plot(X, X_squared, label='X vs X^2', color='blue')
ax1.set_xlabel('X')
ax1.set_ylabel('X^2', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

plt.title('Plotting X vs X^2')
plt.show()
```

```python
fig, ax2 = plt.subplots()
ax2.scatter(X, np.exp(X), color='red', label='X vs exp(X)')
ax2.set_ylabel('exp(X)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('Plotting X vs exp(X)')
plt.show()
```

### 5.	Remove all the symbols (\$\.)from the 'Price' column and convert  and update it to  64-bit int numbers using astype(‘int64’) Hint. Look for string functions like replace/strip).

```python
car_sales['Price'] = car_sales['Price'].str.replace(r'\$|,', '', regex=True).astype('int64')
```

```python
car_sales.head()
```

### 6.	Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column look up the cumsum() function.

```python
car_sales['Total Sales'] = car_sales['Price'].cumsum()
```

```python
car_sales.info()
```

### 7.	Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today). Check out “date_range” function and keep the periods=len(car_sales). 

```python
periods = len(car_sales)

start_date = pd.datetime.today().date()
sale_dates = pd.date_range(start=start_date, periods=periods)

car_sales['Sale Date'] = sale_dates

car_sales.head()
```

### 8.	Use the matplotlib line plot function to plot the 'Price' column versus the 'Total Sales' column (Give proper title and appropriate labels.

```python
plt.plot(car_sales['Price'],car_sales['Total Sales'])

plt.ylabel('Total Sales')
plt.xlabel('Price')
plt.title('Price vs Total Sales Line Plot')

plt.show()
```

### 9.	Create a matplotlib scatter plot of the 'Odometer (KM)' vs 'Price' column.

```python
plt.scatter(car_sales['Odometer (KM)'],car_sales['Price'])

plt.ylabel('Price')
plt.xlabel('Odometer (KM)')
plt.title('Odometer (KM) vs Price Scatter Plot')

plt.show()
```
